```python
import pandas as pd
import os
import subprocess
```

```python
def latexContent(verse1,verse1loc,verse2,verse2loc):
    return f"""\\documentclass{{article}}
\\usepackage[utf8]{{inputenc}}
\\usepackage{{geometry}}
\\usepackage{{setspace}}

% Set the page dimensions and margins
\\geometry{{paperwidth=4.68in, paperheight=6.24in, margin=0.8in}} % Remove margin if you want text to be more freely centered

% Removing page numbers and headers/footers
\\pagestyle{{empty}}

\\begin{{document}}

% Center everything vertically and horizontally
\\noindent
\\topskip0pt
\\vspace*{{\\fill}} % Add vertical space that will be equal on top and bottom
  \\begin{{center}}
    \\normalsize % Font size

    % Your text here
    {verse1}\\\\
    \\textbf{{{verse1loc}}}\\\\
    {verse2}\\\\
    \\textbf{{{verse2loc}}}
  \\end{{center}}
\\vspace*{{\\fill}} % This adds equal space at the bottom, completing the vertical centering

\\end{{document}}
"""

def compile_latex(tex_file):
    # Command to compile LaTeX file to PDF
    compile_command = ['pdflatex', '-interaction=nonstopmode', tex_file]
    try:
        subprocess.run(compile_command, check=True)
        print("PDF generated successfully.")
    except subprocess.CalledProcessError as e:
        print("Failed to compile LaTeX file:", e)

def convert_pdf_to_png(pdf_file, png_file):
    # Command to convert PDF to PNG with specified options
    convert_command = [
        'convert', '-density', '300', pdf_file, 
        '-quality', '100', '-resize', '1404x1872', png_file
    ]
    try:
        subprocess.run(convert_command, check=True)
        print("PNG generated successfully.")
    except subprocess.CalledProcessError as e:
        print("Failed to convert PDF to PNG:", e)
```

```python
df = pd.read_csv('Losungen.csv', sep='\t', header=None, names=['Date', 'Day', 'Extra','verse1loc','verse1','verse2loc','verse2'], skiprows=1, na_values=['NA', 'missing'], encoding='ISO-8859-1')
df = df.replace(to_replace='/', value='', regex=True) # remove all ´/´
```

```python
N = len(df)
date = df.Date.values
verse1 = df.verse1.values
verse1loc = df.verse1loc.values
verse2 = df.verse2.values
verse2loc = df.verse2loc.values

for i in range(N):
    filename = date[i] + '.tex'
    pdf_file = date[i] +'.pdf'
    png_file = date[i] + '.png'
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(latexContent(verse1[i],verse1loc[i],verse2[i],verse2loc[i]))
    compile_latex(filename)
    convert_pdf_to_png(pdf_file,png_file)
os.system('rm *.log *.tex *.aux *.gz *.pdf')
```
